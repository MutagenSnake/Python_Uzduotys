# 9 Parašykite funkciją, kuri mėnesio pavadinimą verstų skaičiumi. Naudodami .apply() lentelėje pakeiskite mėnesių pavadinimus skaičiais.

```python
menesiai = {1: 'January',
  2: 'February',
  3: 'March',
  4: 'April',
  5: 'May',
  6: 'June',
  7: 'July',
  8: 'August',
  9: 'September',
  10: 'October',
  11: 'November',
  12: 'December'}
```

```python
def return_month_number(month):
  all_months = {1: 'January',
  2: 'February',
  3: 'March',
  4: 'April',
  5: 'May',
  6: 'June',
  7: 'July',
  8: 'August',
  9: 'September',
  10: 'October',
  11: 'November',
  12: 'December'}
  for id, month_dict in all_months.items():
        if month == month_dict:
            return id
    
```

```python
return_month_number('November')
```

```python
california['month'].apply(return_month_number)
```

```python
california['month'] = california['month'].apply(return_month_number)
```

# 10 Nuskaitykite lentelę iš [čia](https://lt.wikipedia.org/wiki/S%C4%85ra%C5%A1as:Lietuvos_miestai_pagal_gyventojus). 

```python
wiki = pd.read_html('https://lt.wikipedia.org/wiki/S%C4%85ra%C5%A1as:Lietuvos_miestai_pagal_gyventojus')
```

```python
wiki[0]
```

```python
wiki_cities = wiki[0]
```

```python
wiki_cities.set_index('Miestas', inplace=True)
```

```python
wiki_cities.head()
```

```python
wiki_cities.info()
```

```python
wiki_cities.fillna(0, inplace = True)
```

```python
wiki_cities.info()
```

```python
def turn_to_integer(value):
    return int(value)
```

```python
wiki_cities['1970\xa0m.'] = wiki_cities['1970\xa0m.'].apply(turn_to_integer)
wiki_cities['1959\xa0m.'] = wiki_cities['1959\xa0m.'].apply(turn_to_integer)
```

```python
wiki_cities.info()
```

```python
def fix_object(value):
  if type(value) is int:
    return value
  elif '*' in value:
    clean_value = value.replace('*', '')
    return int(value)
  elif value == '(?)' or value == '(??)':
    return 0
```

```python
wiki_cities['1923\xa0m.'].apply(fix_object)
```

```python
def fix_object_2(value):
  if type(value) is int:
    return 'That was integer'
  elif type(value) is str:
    return 'Tring'
  elif value == '(?)' or value == '(??)':
    return 'Well ???'
  else:
    return "That's else"
```

```python
wiki_cities['1923\xa0m.'].apply(fix_object_2)
```

```python
def fix_object_final(value):
  if type(value) is int:
    return value
  elif type(value) is str:
    if '?' not in value:
      clean_value = value.replace('*', '')
      return int(clean_value)
    elif '?' in value:
      return int(0)
```

```python
wiki_cities['1923\xa0m.'] = wiki_cities['1923\xa0m.'].apply(fix_object_final)
wiki_cities['1897\xa0m.'] = wiki_cities['1897\xa0m.'].apply(fix_object_final)
```
